```python
soup = BeautifulSoup(webpage,'lxml')

# len(soup.find_all('div'))

laptops_name = [laptop.text for laptop in soup.find_all('p', class_= 'Item__name___1fPgt')]
laptops_name
```

```python
laptop_name = []
for item in laptops_name:
    split_values = item.split('/')
    laptop_name.append(split_values)

laptop_name


def extract_elements(laptop_name, index):
    """
    Extract elements at a specific index from each sublist and store them in a new list.

    Args:
    laptops_info (list): The list of sublists containing information about laptops.
    index (int): The index of the element to extract from each sublist.

    Returns:
    list: A list containing the elements extracted from each sublist.
    """
    elements = []
    for info in laptop_name:
        if len(info) > index:
            elements.append(info[index])
        else:
            elements.append(None)  # Append None if the index is out of range for a sublist
    return elements


# Store first elements into one list
first_elements = extract_elements(laptop_name, 0)
# print("First Elements:", first_elements)

# # Store second elements into another list
second_elements = extract_elements(laptop_name, 1)
# print("Second Elements:", second_elements)

# # Store third elements into another list
third_elements = extract_elements(laptop_name, 2)
# print("Third Elements:", third_elements)

# # Store forth elements into another list
forth_elements = extract_elements(laptop_name, 3)
# print("Fhird Elements:", third_elements)



laptops_dimen = [laptop.text for laptop in soup.find_all('ul', class_= 'Item__cardProps___Hxy-F')]
laptops_dimen[0]


laptop_dimension = []
for item in laptops_dimen:
    split_values = item.split('\n')
    laptop_dimension.append(split_values)

laptop_dimension


def extract_dimension(laptop_dimension, index):
    """
    Extract elements at a specific index from each sublist and store them in a new list.

    Args:
    laptops_info (list): The list of sublists containing information about laptops.
    index (int): The index of the element to extract from each sublist.

    Returns:
    list: A list containing the elements extracted from each sublist.
    """
    elements = []
    for info in laptop_dimension:
        if len(info) > index:
            elements.append(info[index])
        else:
            elements.append(None)  # Append None if the index is out of range for a sublist
    return elements


# Store first elements into one list
screen_dimen = extract_dimension(laptop_dimension, 4)
# print("First Elements:", screen_dimen)

# # Store second elements into another list
weight = extract_dimension(laptop_dimension, 8)
# print("Second Elements:", weight)




laptop_specification = {'laptop name':first_elements,
                        'GPU':second_elements,
                        'Main Memory':third_elements, 
                        'Secondary Memory':forth_elements, 
                        'Screen Dimension':screen_dimen,
                        'Weight': weight}
df = pd.DataFrame(laptop_specification)
df


final_df = df.to_csv('final_df.csv', index = True)
```
