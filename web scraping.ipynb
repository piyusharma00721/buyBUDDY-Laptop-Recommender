```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

def extract_laptop_data(url):
    """Extracts laptop data from a given URL and returns it as a DataFrame.

    Args:
        url (str): The URL of the webpage containing laptop information.

    Returns:
        pd.DataFrame: A DataFrame containing the extracted laptop data.
    """

    webpage = requests.get(url).text
    soup = BeautifulSoup(webpage, 'lxml')

    laptops_name = [laptop.text.strip() for laptop in soup.find_all('p', class_= 'Item__name___1fPgt')]

    laptops_dimen = [laptop.text.strip() for laptop in soup.find_all('ul', class_= 'Item__cardProps___Hxy-F')]

    def extract_elements(laptop_, index):
        """Extracts elements at a specific index from each sublist and stores them in a DataFrame column.

        Args:
            laptop_info (list): The list of sublists containing information about laptops.
            index (int): The index of the element to extract from each sublist.

        Returns:
            pd.Series: A Series containing the extracted elements.
        """
        elements = []
        for info in laptop_name:
            if len(info) > index:
                  elements.append(info[index])
            else:
                elements.append(None)  # Append None if the index is out of range for a sublist
        return elements
        
        # elements = pd.Series([info[index] if len(info) > index else None for info in laptop_info])
        # return elements

    # Extract data directly into DataFrame columns using `extract_elements`
    df = pd.DataFrame({
        'laptop name': extract_elements(laptops_name, 0),
        'GPU': extract_elements(laptops_dimen, 1),  # Assuming GPU info is at index 1
        'Main Memory': extract_elements(laptops_dimen, 2),
        'Secondary Memory': extract_elements(laptops_dimen, 3),
        'Screen Dimension': extract_elements(laptops_dimen, 4),
        'Weight': extract_elements(laptops_dimen, 8)
    })

    return df

# Create an empty DataFrame to store all scraped data
final_df = pd.DataFrame()

for j in range(1, 15):
    url = 'https://versus.com/en/laptop?page={}'.format(j)
    print(url)

    # Extract data for each page and append to final DataFrame
    final_df = pd.concat([final_df, extract_laptop_data(url)], ignore_index=True)

# Save the final DataFrame to a CSV file
final_df.to_csv('scraped_laptops.csv', index=False)  # Optionally, specify a filename
print("Data saved to scraped_laptops.csv")
```
