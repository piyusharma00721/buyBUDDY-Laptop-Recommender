```python
# imports neccessary libraries. 
import requests     # request library to make request on website header  for web scrapping.
from bs4 import BeautifulSoup       # beautiful soup  to read web scrapped data.
import pandas as pd         # if you don't know this, get some sleep for forever.

# Initialize an empty DataFrame to store all data for csv
final_df = pd.DataFrame()

for j in range(1, 21):      # itrate whole process 21, i.e., upto 21th page result for laptop data
    url = 'https://versus.com/en/laptop?page={}'.format(j)     
    print(url)
    webpage = requests.get(url).text        # this method used to get all website data infrom on text
    soup = BeautifulSoup(webpage, 'html.parser')        # soup used to read web scrapped data in html parser fromat

    laptops_name = [laptop.text for laptop in soup.find_all('p', class_='Item__name___1fPgt')]          #extract laptop name form soup, with find_all method which return all occurance
    laptop_name = []        
    for item in laptops_name:           # more refining form laptops_name to laptop_name  
        split_values = item.split('/')
        laptop_name.append(split_values)

    def extract_elements(laptop_name, index):
        elements = []
        for info in laptop_name:
            if len(info) > index:
                elements.append(info[index])
            else:
                elements.append(None)  # Append None if the index is out of range for a sublist
        return elements
    #
    first_elements = extract_elements(laptop_name, 0) 
    #print(first_elements)
    second_elements = extract_elements(laptop_name, 1)
    third_elements = extract_elements(laptop_name, 2)
    forth_elements = extract_elements(laptop_name, 3)

    laptops_dimen = [laptop.text for laptop in soup.find_all('ul', class_='Item__cardProps___Hxy-F')]
    laptop_dimension = []
    for item in laptops_dimen:
        split_values = item.split('\n')
        laptop_dimension.append(split_values)

    def extract_dimension(laptop_dimension, index):
        elements = []
        for info in laptop_dimension:
            if len(info) > index:
                elements.append(info[index])
            else:
                elements.append(None)  # Append None if the index is out of range for a sublist
        return elements

    screen_dimen = extract_dimension(laptop_dimension, 4)
    weight = extract_dimension(laptop_dimension, 8)

    laptop_specification = {
        'laptop name': first_elements,
        'GPU': second_elements,
        'Main Memory': third_elements,
        'Secondary Memory': forth_elements,
        'Screen Dimension': screen_dimen,
        'Weight': weight
    }

    # Create a DataFrame from the current laptop_specification dictionary
    current_df = pd.DataFrame(laptop_specification)

    # Append the current DataFrame to the final DataFrame
    final_df = pd.concat([final_df, current_df], ignore_index=True)

# Save the final DataFrame to a CSV file
final_df.to_csv('laptop_specifications.csv', index=False)

print("Data has been successfully saved to laptop_specifications.csv")
```
